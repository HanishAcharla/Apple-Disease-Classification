```python
import os
import shutil
import pandas as pd
```

```python
# Path to the directory containing the images
image_dir = '/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/train_images'
# Path to the directory where you want to create the folders
output_dir = '/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/Dataset_Splitted'
# Path to the CSV file containing the image-to-split mapping
csv_file = '/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/Train_Modified.csv'
# Load the CSV file into a DataFrame
splits = pd.read_csv(csv_file)
```

```python
# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)
# Create folders for each split and move the images
for _, row in splits.iterrows():
    image_name = row['image']  # Column name for image filenames
    split = row['split']  # Column name for split
    # Create the split directory if it doesn't exist
    split_dir = os.path.join(output_dir, split)
    os.makedirs(split_dir, exist_ok=True)
    # Move the image to the corresponding split folder
    src_path = os.path.join(image_dir, image_name)
    dest_path = os.path.join(split_dir, image_name)
    # Check if the source image file exists before moving
    if os.path.exists(src_path):
        shutil.move(src_path, dest_path)
print("Images have been organized into separate folders for each split.")
```

```python
# Path to the directory containing the test images
image_dir = '/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/Dataset_Splitted/Train'
# Path to the directory where you want to create the folders
output_dir = '/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/Final_Dataset/Train'
# Path to the CSV file containing the image-to-label mapping
csv_file = '/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/Train_Modified.csv'
# Load the CSV file into a DataFrame
labels = pd.read_csv(csv_file)
```
