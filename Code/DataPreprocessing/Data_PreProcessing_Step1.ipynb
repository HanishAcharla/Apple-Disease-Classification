```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/train.csv")
data.head()
```

```python
# no of samples that belogs to each class in the dataset
data['labels'].value_counts()
```

```python
#droping the multilabel classes
data_modified= data.drop(data[(data['labels'] == 'scab frog_eye_leaf_spot') | (data['labels'] == 'rust frog_eye_leaf_spot')].index)
```

```python
#combine the complex classes together
complex_index = data_modified[data_modified['labels'].isin(['scab frog_eye_leaf_spot complex','frog_eye_leaf_spot complex','rust complex','powdery_mildew complex'])].index
data_modified.loc[complex_index,'labels']='complex'
```

```python
data_modified['labels'].value_counts()
```

```python
import numpy as np
# Define the splits and their corresponding weights
splits = ['Train', 'Validation', 'Test']
weights = [0.8, 0.1, 0.1]  # Probabilities for each split train = 80%, validation + test = 20%
# Generate random splits for the 'split' column with specified weights
data_modified['split'] = np.random.choice(splits, size=len(data_modified), p=weights)
```

```python
data_modified['split'].value_counts()
```

```python
data_modified.to_csv('/content/drive/MyDrive/1:1_Hanish_Acharla/Dataset/Train_Modified.csv', index=False)
```
