```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
path = "/content/drive/MyDrive/1:1_Hanish_Acharla/Results/FINAL RESNET.csv"
data = pd.read_csv(path)
data.head()
```

```python
#Graph ResNet train accuracies using HPT sheet
epochs = list(data["Epochs/Learning rate"])
lr = list(data.columns)
lr = lr[1:]
all_acc = []
for index in range(len(epochs)):
  all_acc.append(list(data.iloc[index,1:]))
plt.figure(figsize = (9,9))
for index in range(len(epochs)):
  plt.plot(lr, all_acc[index], label = "Epochs: {}".format(epochs[index]))
plt.xlabel("Learning Rate",fontsize=)
plt.ylabel("Accuracy")
#plt.title("Accuracy values with Learning rate and Epochs")
plt.grid(True)
plt.legend(loc = "best")
plt.show()
```
